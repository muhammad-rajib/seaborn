```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
sb.set_theme(color_codes=True)
```

```python
# load dataset
tips = sb.load_dataset('tips')
tips.head(10)
```

### Function to draw linear regressions mdoels

```python
# seaborn.regplot()
sb.regplot(data=tips, x='total_bill', y='tip')
```

```python
# seaborn.lmplot()
sb.lmplot(data=tips, x='total_bill', y='tip')
```

### Fitting different kinds of models

```python
anscombe = sb.load_dataset('anscombe')
anscombe.head(10)
```

```python
sb.lmplot(data=anscombe.query("dataset =='I'"),
         x='x', y='y', ci=None, scatter_kws={'s': 80})
```

```python
# seaborn.residplot()
sb.residplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
              scatter_kws={"s": 80});
```

### Conditioning on other variables

```python
sb.lmplot(data=tips, x='total_bill', y='tip', hue='smoker',
         markers=['o', 'x'], palette='Set1', ci=None)
```

```python
sb.lmplot(data=tips, x='total_bill', y='tip', hue='smoker', 
         col='time', row='sex')
```

### Controlling the size and shape of the plot

```python
f, ax = plt.subplots(figsize=(5, 6))
sb.regplot(x='total_bill', y='tip', data=tips, ax=ax)
```

```python
sb.lmplot(data=tips, x='total_bill', y='tip', 
          col='day', col_wrap=2, height=3)
```

### Plotting a regression in other contexts

```python
sb.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
```

```python
sb.pairplot(data=tips, x_vars=['total_bill', 'size'], y_vars=['tip'],
           height=5, aspect=.8, kind='reg', hue='smoker')
```
