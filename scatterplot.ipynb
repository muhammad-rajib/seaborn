## seaborn.scatterplot

```python
import seaborn as sb

sb.set_theme(style='darkgrid')
```

```python
# load dataset
tips = sb.load_dataset('tips')
tips.head(10)
```

```python
sb.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
sb.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
```

```python
sb.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='time')
```

```python
sb.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
```

```python
sb.scatterplot(data=tips, x='total_bill', y='tip', hue='size')
```

```python
# change color of plots and tip rate
sb.scatterplot(data=tips, x='total_bill', y='tip', hue='size', palette='deep')
```

```python
tip_rate = tips.eval('tip / total_bill').rename('tip_rate')
sb.scatterplot(data=tips, x='total_bill', y='tip', hue=tip_rate)
```

```python
sb.scatterplot(data=tips, x='total_bill', y='tip', hue='size',
               size='size', sizes=(20, 200), legend='full')
```

```python
sb.scatterplot(data=tips, x='total_bill', y='tip', hue='size',
               size='size', sizes=(20, 200), hue_norm=(0, 7), legend='full')
```

```python
markers = {'Lunch':'s', 'Dinner':'X'}
sb.scatterplot(data=tips, x='total_bill', y='tip', hue='day', 
               style='time', markers=markers)
```

```python
markers = {'Lunch':'s', 'Dinner':'X'}
sb.relplot(data=tips, x='total_bill', y='tip', hue='day', 
               style='time', markers=markers, col='sex')
```
