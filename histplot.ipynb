## seaborn.histplot

```python
import seaborn as sb
import pandas as pd
```

```python
# load dataset
penguins = sb.load_dataset("penguins")
penguins.head(10)
```

```python
sb.histplot(data=penguins, x='flipper_length_mm')
```

```python
sb.histplot(data=penguins, y='flipper_length_mm')
```

```python
sb.histplot(data=penguins, x='flipper_length_mm', binwidth=3)
```

```python
sb.histplot(data=penguins, x='flipper_length_mm', bins=20)
```

```python
sb.histplot(data=penguins, x='flipper_length_mm', kde=True)
```

```python
# Multiple histograms from a long-form dataset with hue mapping
sb.histplot(data=penguins, x='flipper_length_mm', hue='species', multiple='stack')
```

```python
# when using a hue sementic with discrete data, it can make sense to 'dodge' the levels
sb.histplot(data=penguins, x='flipper_length_mm', hue='species', multiple='dodge')
```

```python
# unfilled the bars
sb.histplot(data=penguins, x='flipper_length_mm', hue='species', multiple='dodge', fill=False)
```

```python
# load 'planet' dataset
planets = sb.load_dataset('planets')
planets.head(10)
```

```python
sb.histplot(
    data=planets, x='distance', hue='method',
    hue_order=['Radial Velocity', 'Transit'], 
    log_scale=True, element='step', fill=False,
    cumulative=True, stat='density', common_norm=False
)
```

```python
# add colorbar
sb.histplot(
    data=planets, x='year', y='distance',
    bins=30, discrete=(True, False), log_scale=(False, True),
    cbar=True, cbar_kws=dict(shrink=.75)
)
```
