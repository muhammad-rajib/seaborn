## seaborn.lineplot

```python
import seaborn as sb

sb.set_theme(style='darkgrid')
```

```python
# load dataset
flights = sb.load_dataset('flights')
flights.head(15)
```

```python
# may_flights = flights[(flights['month'] == 'May')]
may_flights = flights.query("month == 'May'")
may_flights
```

```python
sb.lineplot(data=may_flights, x='year', y='passengers')
```

```python
# let's use wide-form data
flights_wide = flights.pivot("year", "month", "passengers")
flights_wide
```

```python
sb.lineplot(data=flights_wide['May'])
```

```python
sb.lineplot(data=flights_wide)
```

```python
sb.relplot(data=flights_wide, kind='line')
```

```python
sb.lineplot(data=flights, x='year', y='passengers', hue='month', style='month')
```

```python
# load fmri dataset
fmri = sb.load_dataset('fmri')
fmri.head(10)
```

```python
sb.lineplot(
    data=fmri,
    x='timepoint', y='signal', hue="event", style="event",
    markers=True, dashes=False
)
```

```python
sb.lineplot(
    data=fmri,
    x='timepoint', y='signal', hue='event', err_style='bars', ci=68
)
```

```python
# load seaborn 'dots' dataset
dots = sb.load_dataset('dots').query("align == 'dots'")
dots.head()
```

```python
sb.lineplot(
    data=dots.query("coherence > 0"),
    x='time', y='firing_rate',
    hue='coherence', style='choice',
    palette='flare'
)
```

```python
import matplotlib as mpl
sb.lineplot(
    data=dots.query("coherence > 0"),
    x='time', y='firing_rate',
    hue='coherence', style='choice',
    palette='flare', hue_norm=mpl.colors.LogNorm()
)
```

```python
palette = sb.color_palette('mako_r', 6)
sb.lineplot(
    data=dots, x='time', y='firing_rate',
    hue='coherence', style='choice',
    palette=palette
)
```

```python
sb.lineplot(
    data=dots, x='time', y='firing_rate',
    hue='choice', size='coherence', sizes=(.25, 2.5)
)
```
