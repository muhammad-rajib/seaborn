## seaborn.rugplot

```python
import seaborn as sb

sb.set_theme()
```

```python
# load dataset
tips = sb.load_dataset('tips')
tips.head(10)
```

```python
sb.displot(data=tips, x='total_bill', kde=True)
```

```python
sb.kdeplot(data=tips, x='total_bill')
sb.rugplot(data=tips, x='total_bill')
```

```python
sb.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
sb.rugplot(data=tips, x='total_bill', y='tip', hue='time')
```

```python
sb.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
sb.rugplot(data=tips, x='total_bill', y='tip', hue='time', 
            height=.1)
```

```python
sb.scatterplot(data=tips, x='total_bill', y='tip')
sb.rugplot(data=tips, x='total_bill', y='tip', height=-.02, clip_on=False)
```
