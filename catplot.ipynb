## seaborn.catplot
> Default plots: kind=stripplot

```python
import seaborn as sb
import matplotlib.pyplot as plt

sb.set_theme(style='ticks', color_codes=True)
```

```python
# load dataset
tips = sb.load_dataset('tips')
tips.head(10)
```

```python
sb.catplot(data=tips, x='day', y='total_bill') # default: jitter=True
```

```python
sb.catplot(data=tips, x='day', y='total_bill', jitter=False)
```

### seaborn.swarmplot 
> Draw a categorical scatterplot with non-overlapping points.

```python
# set kind='swarmplot'
sb.catplot(data=tips, x='day', y='total_bill', kind='swarm')
```

```python
# set hue semantic
sb.catplot(data=tips, y='total_bill', x='day', kind='swarm', hue='sex')
```

```python
sb.catplot(data=tips, x='smoker', y='tip', order=['No', 'Yes'])
```

## Boxplots

```python
sb.catplot(data=tips, x='day', y='total_bill', kind='box', hue='smoker')
```

## Violinplots

```python
sb.catplot(data=tips, x='total_bill', y='day', hue='sex')
```

```python
sb.catplot(data=tips, x='total_bill', y='day', hue='sex', 
           kind='violin', cut=0, bw=.15)
```

```python
sb.catplot(data=tips, x='day', y='total_bill', hue='sex',
          kind='violin', split=True, inner='stick', palette='pastel'
)
```

```python
sb.catplot(data=tips, x='day', y='total_bill', kind='violin', inner=None)
sb.swarmplot(data=tips, x='day', y='total_bill', color='k', size=3)
```

### Showing multiple relationships with facets

```python
sb.catplot(data=tips, x='day', y='total_bill', hue='smoker', 
           col='time', kind='swarm', aspect=.7
)
```

```python
titanic = sb.load_dataset('titanic')
g = sb.catplot(data=titanic.query('fare > 0'), x='fare', y='survived', 
           row='class', kind='box', height=1.5, aspect=4, orient='h'
)

g.set(xscale='log')
```

### Points plots

```python
sb.catplot(data=titanic, x='class', y='survived', hue='sex',
          kind='point', linestyles=['-', '--'], markers=['^', 'o'],
          palette={'male': 'm', 'female': 'c'} 
)
```
