## seaborn.displot
> Figure-level interface for drawing distribution plots onto a FcetGrid

    histplot() (with kind="hist"; the default)
    kdeplot() (with kind="kde")
    ecdfplot() (with kind="ecdf"; univariate-only)

```python
# import libraries
import seaborn as sb

sb.set_theme(style='darkgrid')
```

```python
penguins = sb.load_dataset('penguins')
penguins.head()
```

```python
sb.displot(data=penguins, x='flipper_length_mm')
```

```python
# kind='kde', kde=kerner den
sb.displot(data=penguins, x='flipper_length_mm', kind='kde')
```

```python
sb.displot(data=penguins, x='flipper_length_mm', kind='ecdf')
```

```python
sb.displot(data=penguins, x='flipper_length_mm', kde=True)
```

```python
sb.displot(data=penguins, x='flipper_length_mm', hue='species', kind='kde')
```

```python
sb.displot(data=penguins, x='flipper_length_mm', hue='species', multiple='stack')
```

```python
sb.displot(data=penguins, x='flipper_length_mm', hue='species', col='species')
```

```python
sb.displot(data=penguins, x='flipper_length_mm', hue='species', col='sex', kind='kde')
```

```python
sb.displot(
    data=penguins, y='flipper_length_mm', hue="sex",
    col='species', height=4, aspect=.7, kind='ecdf'
)
```

```python
g = sb.displot(
    data=penguins, y="flipper_length_mm", hue="sex", col="species",
    kind="kde", height=4, aspect=.7
)

g.set_axis_labels("Density (a.u.)", "Flipper length (mm)")
g.set_titles("{col_name} penguins")
```
