## seaborn.relplot
> Figure-level interface for drawing relational plots onto a FcetGrid

```python
# import libraries
import seaborn as sb

sb.set_theme(style='darkgrid')
%matplotlib inline
```

```python
# load dataset
tips = sb.load_dataset('tips')
tips.head(10)
```

```python
sb.relplot(data=tips, x='total_bill', y='tip', hue='day')
```

```python
sb.relplot(data=tips, x='total_bill', y='tip', hue='day', col='time',
          style='day')
```

```python
sb.relplot(data=tips, x='total_bill', y='tip', hue='day',
          col='time', row='sex')
```

```python
sb.relplot(data=tips, x='total_bill', y='tip', hue='time',
          col='day', col_wrap=2)
```

```python
sb.relplot(
    data=tips, x='total_bill', y='tip',
    hue='time', size='size', style='sex',
    palette=['orange', 'c'], sizes=(10, 100)
)
```

### seaborn.relplot()
> set kind="line" to draw the figure using lineplot() function

```python
# dataset
fmri = sb.load_dataset('fmri')
fmri.head(10)
```

```python
g = sb.relplot(data=fmri, x='timepoint', y='signal', hue='event',
          col='region', style='event', height=4, 
           aspect=.7, kind='line')

g.set_axis_labels('Timepoint', 'Percent signal change')
g.set_titles('Region: {col_name} cortex')
g.tight_layout(w_pad=0)
```

```python
# using wide-form data
flights = sb.load_dataset('flights').pivot("year", "month", "passengers")
flights.head(10)

sb.relplot(data=flights, kind='line')
```
