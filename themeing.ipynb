```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
sb.barplot(x=['A', 'B', 'C'], y=[1, 3, 2])
```

```python
# set theme
sb.set_theme()
sb.barplot(x=['A', 'B', 'C'], y=[1, 3, 2])
```

```python
sb.set_theme(style='whitegrid', palette='pastel')
sb.barplot(x=['A', 'B', 'C'], y=[1, 3, 2])
```

```python
custom_params = {'axes.spines.right': False, 'axes.spines.top': False,
                #'xtick.bottom': False,
}
sb.set_theme(style='ticks', rc=custom_params)
sb.barplot(x=['A', 'B', 'C'], y=[1, 3, 2])
```

```python
# temporarily change the style of plots
with sb.axes_style("whitegrid"):
    sb.barplot(x=['A', 'B', 'C'], y=[1, 3, 2])
```

```python
sb.barplot(x=['A', 'B', 'C'], y=[1, 3, 2])
```

```python
# lets see the predefined style to shows plot values
sb.plotting_context("talk")
```

```python
sb.set_theme(style='ticks')
with sb.plotting_context("talk"):
    sb.lineplot(x=['A', 'B', 'C'], y=[1, 3, 2])
```

```python
# set_context()
sb.set_context('notebook')
sb.lineplot(x=[0, 1, 2], y=[1, 3, 2])
```

```python
sb.set_context('notebook', font_scale=1.50, rc={'lines.linewidth': 3})
sb.lineplot(x=[0, 1, 2], y=[1, 3, 2])
```

```python
# seaborn.reset_defaults
sb.reset_defaults()
sb.lineplot(x=[0, 1, 2], y=[1, 3, 2])
```
